```python
!pip install seaborn==0.9.0

import seaborn as sns
print(sns.__version__)

import pandas as pd

import statistics as stats

import matplotlib.pyplot as plt

# Lendo o arquivo do excel
desafio = pd.read_excel("/content/DadosDesafio.xlsx")
```

```python
desafio.head()
#Exibindo uma amostra do arquivo
```

```python
desafio
#Exibindo todo o conteud do arquivo
```

```python
# Exibindo as medidas de tendência central para o: Aluguel
print("Média do valor do aluguel em R$:",desafio["Aluguel"].mean().round(2))
print("Valor Máximo do aluguel em R$:",desafio["Aluguel"].max())
print("Valor Mínimo do aluguel em R$:",desafio["Aluguel"].min())
print("Valor Mediano do aluguel em R$:",desafio["Aluguel"].median())
print("Valor da Moda do aluguel em R$:",stats.multimode(desafio["Aluguel"]))
print("Valor do Desvio padrão do aluguel:",desafio["Aluguel"].std().round(2))
```

```python
# Exibindo as medidas de tendência central para o: Juros
print("Média do valor dos juros em %:",desafio["Juros (%)"].mean().round(2))
print("Valor máximo dos juros em %:",desafio["Juros (%)"].max())
print("Valor mínimo dos juros em %:",desafio["Juros (%)"].min())
print("Valor Mediano dos juros em %:",desafio["Juros (%)"].median())
print("Valor da moda dos juros em %:",stats.multimode(desafio["Juros (%)"]))
print("Valor do Desvio padrão dos juros:",desafio["Juros (%)"].std().round(2))
```

```python
# Exibindo as medidas de tendência central para a: Data de Vencimento
print("Média do dia do vencimento:",desafio["Dia Vcto"].mean().round(2))
print("Valor máximo do dia do vencimento:",desafio["Dia Vcto"].max())
print("Valor mínimo do dia do vencimento:",desafio["Dia Vcto"].min())
print("Valor Mediano do dia do vencimento:",desafio["Dia Vcto"].median())
print("Valor da moda do dia do vencimento:",stats.mode(desafio["Dia Vcto"]))
print("Valor do Desvio padrão do dia do vencimento:",desafio["Dia Vcto"].std().round(2))
```
