```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine.csv', sep=",", header='infer', encoding='utf-8')
```

```python
df.head()
```

```python
df.columns=['Álcool', 'Ácido Malico', 'Cinzas', 'Alcalinidade das cinzas', 'Magnésio','Fenóis Totais','Flavonoides','Fenóis não-flavonoides','Proantocianinas', 'Intensidade da Cor', 'Matiz', 'OD280_OD315', 'Prolina']
df.head()
```

```python
fig = px.parallel_coordinates(df, color=df['Intensidade da Cor'],
                              dimensions=['Álcool','Ácido Malico','Alcalinidade das cinzas','Magnésio','Fenóis Totais','Flavonoides', 'Fenóis não-flavonoides'],
                              #dimensions=['Alcohol', 'Malic_acid',	'Ash',	'Alcalinity_ash',	'Magnesium',	'Total_phenols',	'Flavanoids',	'Nonflavanoid_phenols'],
                              #color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_scale=px.colors.sequential.Viridis,
                              color_continuous_midpoint=7)
fig.show()
```

```python
fig = px.scatter(df, x="Ácido Malico", y="Álcool", color='Intensidade da Cor',
                 size='Flavonoides', hover_data=['Intensidade da Cor'],
                 color_continuous_scale=px.colors.sequential.Viridis)
fig.show()
```
